```python
# Program to recommend fertilizer based on nitrogen level

N = float(input("Enter soil nitrogen level (mg/kg): "))

if N < 100:
    print("Apply high nitrogen fertilizer.\n"
          "Cost = ₹2000 per bag.\n"
          "Your soil has low nitrogen.")

elif N >= 100 and N <= 250:
    print("Apply medium nitrogen fertilizer.\n"
          "Cost = ₹1000 per bag.\n"
          "Your soil has adequate nitrogen.")

else:
    print("No fertilizer is required.\n"
          "Cost = ₹0\n"
          "Soil is healthy.")
```

```python

# Simple Pest Severity Program

pests = 12
damage = 18
temp = 28
humidity = 75

print("Pests per plant:", pests)
print("Damage:", damage, "%")
print("Temperature:", temp, "°C")
print("Humidity:", humidity, "%")

if pests < 5 and damage < 10:
    print("\nSeverity Level: LOW")
    print("Recommendation: No action needed")
    print("Cost of intervention: ₹0/hectare")

elif pests <= 15 or damage <= 25:
    print("\nSeverity Level: MEDIUM")
    print("Recommendation: Monitor daily. Spray pesticide if needed")
    print("Cost of intervention: ₹500/hectare")

else:
    print("\nSeverity Level: HIGH")
    print("Recommendation: Immediate pesticide spray required")
    print("Cost of intervention: ₹1200/hectare")
```

```python
# Simple Harvest Readiness Program

crop = "Wheat"
days = 165

print("Crop:", crop)
print("Days since sowing:", days)

if crop == "Rice":
    min = 120
    max = 150

elif crop == "Wheat":
    min = 140
    max = 180

elif crop == "Cotton":
    min = 160
    max = 200

if days >= min and days <= max:
    print("\nStatus: READY FOR HARVEST")
    print("Maturity range:", min, "-", max, "days")
    print("Recommendation: Start harvesting within 2 weeks")
    print("Yield loss if delayed: 2% per day")
else:
    print("\nStatus: NOT READY FOR HARVEST")
    print("Maturity range:", min, "-", max, "days")
```

```python

# Base yield
yield_value = 40

# Inputs
rainfall = int(input("Rainfall (mm): "))
temperature = int(input("Temperature (°C): "))
soil = int(input("Soil quality (80-100): "))

# Rainfall effect
if rainfall > 600:
    yield_value += 2
elif rainfall < 400:
    yield_value -= 2

# Temperature effect
if temperature < 25 or temperature > 30:
    yield_value -= 1

# Soil quality effect
yield_value = yield_value * soil / 100

# Output
print("Predicted yield:", yield_value, "quintals/hectare")
```
