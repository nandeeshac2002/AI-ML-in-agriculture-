```python

sentence = "I like tea"

words = sentence.split()

vowels = "aeiouAEIOU"

for word in words:
    count = 0
    for letter in word:
        if letter in vowels:
            count = count + 1
    print(word, "=", count, "vowels")
```

```python

text = "HELLO"


reverse_text = text[::-1]


if text == reverse_text:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
string = input("Enter a string: ")

if string == string[::-1]:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
word = "beautiful"
vowels = "aeiou"
count = 0

for char in word:
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```
