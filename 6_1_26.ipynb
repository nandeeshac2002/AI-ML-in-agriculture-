```python
string = input("Enter a string: ")

if string == string[::-1]:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
word = "beautiful"
vowels = "aeiou"
count = 0

for char in word:
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```
