```python
file = open("daily_expenses.txt", "w")

print("Daily Expenses Sheet\n")

file.write("Daily Expenses Sheet\n")
file.write("--------------------\n")

file.write("Coffee  : 10 Rs\n")
file.write("Tiffen  : 50 Rs\n")
file.write("Dinner  : 50 Rs\n")

total = 10 + 50 + 50
file.write("--------------------\n")
file.write("Total   : " + str(total) + " Rs\n")

file.close()

print("Expenses saved in file successfully")
```
